Task 3 :  Prediction using Decision Tree Algorithm

```
#Essential Libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import sklearn.datasets as datasets
```

```
#Loading datasets
iris=datasets.load_iris()

df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df.head(5))
```

```
#Splitting datasets into features and target 
X=df
y=iris.target
print(y)
```

```
#split data into training data and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```
#Train Decision tree classifier model
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
print("Decison tree classifier is created")
```

```
#Visualization of Decision tree

from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus

dot_data = StringIO()
export_graphviz(tree,out_file = dot_data,filled = True, 
                rounded = True,special_characters = True,feature_names=iris.feature_names,class_names = ['0','1','2'])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('iris.png')
Image(graph.create_png())
```

```
#prediction for the test dataset
y_pred = tree.predict(X_test)

#Comparision between predict and actual class
df = pd.DataFrame({'Predict class':y_pred,'Actual class':y_test})
print(df.head(10))
```

```
#Model Accuracy
print("Accuracy of Decison tree classifier:",metrics.accuracy_score(y_test,y_pred))
```

```
#Confusion matrix and report
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.classification_report(y_test,y_pred))
```
